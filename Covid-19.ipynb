```python
#Cases only in India
India_case=covid[covid["location"]=="India"]
India_case
```

```python
India_case.head()
```

```python
India_case.tail()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Total cases per day
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=India_case)
plt.show()
```

```python
#Making a dataframe for last 5 days
India_last_5_days=India_case.tail()
```

```python
#Total cases in last 5 days
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_cases",data=India_last_5_days)
plt.show()
```

```python
#Total tests per day
sns.set(rc={'figure.figsize':(15,10)})
sns.lineplot(x="date",y="total_tests",data=India_case)
plt.show()
```

```python
#Understanding cases of India, China , Japan and United States
india_japan_china_us=covid[(covid["location"] =="India") | (covid["location"] =="China") | (covid["location"]=="Japan")|(covid["location"]=="United States")]
```

```python
#Plotting growth of cases across China, India and Japan
sns.set(rc={'figure.figsize':(25,20)})
sns.barplot(x="location",y="total_cases",data=india_japan_china_us,hue="date")
plt.show()
```

```python
#Getting latest data
last_day_cases=covid[covid["date"]=="2020-05-25"]
last_day_cases
```

```python
#Sorting data w.r.t total_cases
max_cases_country=last_day_cases.sort_values(by="total_cases",ascending=False)
max_cases_country
```

```python
#Top 5 countries with maximum cases
max_cases_country[1:6]
```

```python
#Sorting data w.r.t total_cases
min_cases_country=last_day_cases.sort_values(by="total_cases",ascending=True)
min_cases_country
```

```python
#Top 5 countries with manimum cases
min_cases_country[1:6]
```

```python
#Making bar-plot for countries with top cases
sns.barplot(x="location",y="total_cases",data=max_cases_country[1:6],hue="location")
plt.show()
```
