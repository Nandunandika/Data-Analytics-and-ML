```

#Splitting the dataset intro training and testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(iris_dataset_features, iris_dataset_labels, test_size=0.30, random_state=42)

import numpy as np

print("Training data is:\n", x_train,"\n")
print("Shape of training data is: ",np.shape(x_train),"\n")
print("Training data labels are:\n", y_train,"\n")
print("Shape of training data labels is: \n", np.shape(y_train))

print("Testing data is:\n", x_test,"\n")
print("Shape of testing data is: ",np.shape(x_test),"\n")
print("Testing data labels are:\n", y_test,"\n")
print("Shape of testing data labels is: \n", np.shape(y_test),"\n")

```

```
#Building the Support Vector Machine classifier

from sklearn.svm import SVC
svm_classifier = SVC(C=5.0, kernel = 'linear', gamma = 'scale')

print(svm_classifier)

#Training the SVM using training data - label pairs

svm_classifier.fit(x_train, y_train)
```

```

#Displaying the SVM classifier arguments

print("Support vectors are: \n", svm_classifier.support_vectors_, "\n")
print("Shape of Support vectors obtained is: ", np.shape(svm_classifier.support_vectors_),"\n")
print("Number of support vectors for each class = ", svm_classifier.n_support_,"\n")
print("Constants in the decision function = ", svm_classifier.intercept_,"\n")
```

```

#Making the predictions on the test dataset

svm_predictions = svm_classifier.predict(x_test)
print("Predictions on test data are: \n", svm_predictions,"\n")

```

```
#Analyzing the model performance

print("Comparing the predictions with the ground-truth:\n", np.column_stack((svm_predictions,y_test)),"\n\n")

accuracy = svm_classifier.score(x_test, y_test)

print("Accuarcy score of the model is: ", accuracy, "\n")

```

```
#Creating the confusion matrix

from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, svm_predictions)

print("Confusion matrix is: \n", cm, "\n")

```

```
#Using PCA for dimensionality reduction

from sklearn.decomposition import PCA

pca = PCA(n_components=2, svd_solver='auto', tol=0.05, iterated_power='auto', random_state=42)
print(pca)

```

```
#Reducing the dimensionality of training data

x_train_pca = pca.fit_transform(x_train)

#Displaying the PCA arguments

print("Principal components in the feature space = ", pca.components_)
print("Explained Variance = ", pca.explained_variance_)
print("Explained Variance Ratio = ", pca.explained_variance_ratio_)
print("Singular Values corresponding to each components = ", pca.singular_values_)

print("Transformed training data is:\n", x_train_pca,"\n")
print("Shape of transformed training data is: \n", np.shape(x_train_pca),"\n")

```

```
#Buiding the SVM classifier

from sklearn.svm import SVC
svm_classifier_pca = SVC(C=5.0, kernel = 'linear', gamma = 'scale')
print(svm_classifier_pca)

#Training the SVm using transformed training data and original labels

svm_classifier_pca.fit(x_train_pca, y_train)

```

```
#Displaying the SVm arguments

print("printing the support vectors ", svm_classifier_pca.support_vectors_,"\n")
print("Number of support vectors for each class = ", svm_classifier_pca.n_support_,"\n")
print("Constants in the decision function = ", svm_classifier_pca.intercept_,"\n")

#Making predictions on the test data

x_test_pca = pca.transform(x_test)
print("Transformed test data is: \n", x_test_pca,"\n")
print("Shape of the transformed test data is: \n", np.shape(x_test_pca),"\n")

svm_predictions_pca = svm_classifier_pca.predict(x_test_pca)

```

```
#Analyzing the predictions 

print("Predictions on the transformed test data  is:\n", svm_predictions_pca,"\n")
print("Comaparing the results againts the ground-truth: \n", np.column_stack((svm_predictions_pca,y_test)),"\n\n")

accuracy_svm = svm_classifier_pca.score(x_test_pca, y_test)
print("Accuracy of the model = ",accuracy_svm)

#Creating the confusion matrix

from sklearn.metrics import confusion_matrix 
cm_pca = confusion_matrix(y_test, svm_predictions_pca)

print("Confusion matrix using PCA is: \n", cm_pca, "\n")
```
